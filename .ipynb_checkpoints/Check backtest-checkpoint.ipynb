```python
from urllib.request import urlopen
import json
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# get data from 1 coin
ticker = 'BTC_ETH'
start_date = 1405699200
end_date = 9999999999
period = 14400

ret = urlopen('https://poloniex.com/public?command=returnChartData&currencyPair={}&start={}&end={}&period={}'.format(ticker, start_date, end_date, period))
raw_chart_data = json.loads(ret.read())
```

```python
df_chart_data = pd.DataFrame(raw_chart_data)
```

```python
df_chart_data['date'] = pd.to_datetime(df_chart_data['date'],unit='s')
print(df_chart_data.shape)
print(df_chart_data.head(10))

df_chart_data['close'].plot()
```

# Backtesting simple strategy

```python
from backtesting import Backtest, Strategy
from backtesting.lib import resample_apply
```

```python
def SMA(array, n):
    """Simple moving average"""
    return pd.Series(array).rolling(n).mean()


def RSI(array, n):
    """Relative strength index"""
    # Approximate; good enough
    gain = pd.Series(array).diff()
    loss = gain.copy()
    gain[gain < 0] = 0
    loss[loss > 0] = 0
    rs = gain.ewm(n).mean() / loss.abs().ewm(n).mean()
    return 100 - 100 / (1 + rs)
```

```python
class System(Strategy):
    d_rsi = 30  # Daily RSI lookback periods
    level = 50
    
    def init(self):
        # Compute moving averages the strategy demands
        self.ma10 = self.I(SMA, self.data.Close, 10)
        self.ma50 = self.I(SMA, self.data.Close, 50)
        self.ma100 = self.I(SMA, self.data.Close, 100)

        
        # Compute daily RSI(30)
        self.daily_rsi = self.I(RSI, self.data.Close, self.d_rsi)
        
                
    def next(self):
        price = self.data.Close[-1]
        
        # If we don't already have a position, and
        # if all conditions are satisfied, enter long.
        if (not self.position and
            self.daily_rsi[-1] > self.level and
            self.ma10[-1] > self.ma50[-1] and
            self.ma50[-1] > self.ma100[-1] and
            price > self.ma10[-1]):
            
            # Buy at market price on next open, but do
            # set 8% fixed stop loss.
            self.buy(sl=.92 * price)
        
        # If the price closes 2% or more below 10-day MA
        # close the position, if any.
        elif price < .98 * self.ma10[-1]:
            self.position.close()
```

```python
#convert columns for backtest
cols = ['open', 'high', 'low', 'close', 'volume', 'date']
df_backtest = df_chart_data[cols]
df_backtest.columns = ['Open', 'High', 'Low', 'Close', 'Volume', 'Date']

# filter dates before 2017
df_backtest.set_index('Date', inplace=True)
df_backtest = df_backtest.loc['2017-09-01 00:00:00':]

print(df_backtest.shape)
df_backtest.head()
```

```python
df_backtest.tail()
```

```python
bt = Backtest(df_backtest, System, cash=10000, commission=.0025)

bt.run()
```
