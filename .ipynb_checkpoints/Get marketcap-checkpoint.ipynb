```python
from urllib.request import urlopen
import json
import pandas as pd
import pickle
```

```python
# get  data
endpoint = "https://api.coingecko.com/api/v3"
link = "/coins/markets?vs_currency=usd&order=market_cap_desc&per_page=300&page=1&sparkline=false"
ret = urlopen(endpoint + link)
marketcap_data = json.loads(ret.read())
```

```python
coins_name = []
coins_symbol = []
market_cap = []
market_cap_rank =[]
circulating_supply = []
total_supply = []

for coin in marketcap_data:
    coins_name.append(coin['name'])
    coins_symbol.append(coin['symbol'])
    market_cap.append(coin['market_cap'])
    market_cap_rank.append(coin['market_cap_rank'])
    circulating_supply.append(coin['circulating_supply'])
    total_supply.append(coin['total_supply'])  
```

```python
data = [coins_name, coins_symbol, market_cap, market_cap_rank, circulating_supply, total_supply]
columns =  ['coins_name', 'coins_symbol', 'market_cap', 'market_cap_rank', 'circulating_supply', 'total_supply']
market_cap_df = pd.DataFrame()

for data_point, column in zip(data, columns):
    market_cap_df[column] = data_point

#market_cap_df = pd.DataFrame(data, columns=columns)
```

```python
print(market_cap_df)
```

```python
# store as obj
with open(r'obj/Coin_Market_data.pkl', "wb") as f:
    pickle.dump(market_cap_df, f)
```
